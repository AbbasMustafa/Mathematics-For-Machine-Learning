# What is Vector?

Vectors are everywhere: physics, engineering, mathematics, computer science, video games, and more. Each field's interpretation of what a vector is may be different, but vectors live a similar life in every space.

1. For physicists, a vector is an "arrow" of a given length (magnitude) and direction. It can represent directional quantities like velocity, force, acceleration.
2. For computer scientists, a vector is an ordered list of numbers. It can represent a set of variables or features stored in order.
3. For mathematicians, vectors are generic objects that behave a certain way when they are added or scaled: $u+v, αv$.

In computer science and in data science, vectors are often multi-dimensional, that is, they have many components. They contain a set of ordered variables in a data model, like for example: the age, weight, daily hours of sleep, weekly hours of exercise, and blood pressure of an individual (five dimensions).

# Fundamental Vector Operation

Two operations are the foundation of everything: **vector addition**, and **multiplication by a scalar** (i.e., scaling).

Lets first vizulaize vector **addition**, suppose we have two vector  $\begin{pmatrix}
1 \\
3 
\end{pmatrix}$ + $\begin{pmatrix}
3 \\
1 
\end{pmatrix}$
When we add them both our resultant vector will be  $\begin{pmatrix}
4 \\
4 
\end{pmatrix}$

**Note**: One important Rule of vector addition is that both vector/matrix must have same dimension.

```python
a = np.array((1,3))
b = np.array((3,1))
origin = np.array((0,0))

vectors = [a, b, a+b]
tails = [origin, a, origin]

plot_vector(vectors, tails)
```

Let's now look at **multiplication by a scalar**: essentially, the length of the vector is scaled by the scalar factor. If you multiply a vector by 2, its length (magnitude) doubles.

For example, if we scale by 2 the vector $c = \begin{pmatrix}
2 \\
2 
\end{pmatrix}$, it looks like this:

$2*\begin{pmatrix}
2 \\
2 
\end{pmatrix} = \begin{pmatrix}
2*2 \\
2*2 
\end{pmatrix} = \begin{pmatrix}
4 \\
4 
\end{pmatrix}$

```python
c = np.array((2,2))
vectors = [c, 2*c]

plot_vector(vectors)
```

# Basis Vectors

With the idea of vector addition and multiplication by scalar in your mind now you can make any vector by these two operations, You just need Basis vectors!

**Basis vector** is a set of n vectors that:
1. Are not linear Comibnation of each others (Linearly independant)
2. Span the Space

So what is **Linearly independant**?<br>
Those vectors are Linearly independant where you can't make one vector from another just like:
$\begin{pmatrix}
1 \\
0 
\end{pmatrix}$ $\begin{pmatrix}
0 \\
1 
\end{pmatrix}$

what is **Span the space**?<br>
Span of basis vectors is where you can make another vector with some kind of addiition and scalar multiplcation, for 2D-plane basis vectors have all span of 2D-Plane.

```python
i = np.array((1,0))
j = np.array((0,1))

vec = 3*i + 2*j
vectors = [i, j , 3*i , 2*j, vec]

plot_vector(vectors)
```

## Linear Combination

Adding two vectors that were each multiplied by a scalar is called a linear combination of those two vectors.This brings us to the idea of the span of two vectors: the set of all possible linear combinations of the two.

```python
from numpy.random import randint

vectors = []

for _ in range(300):
    m = randint(-10,10)
    n = randint(-10,10)
    vectors.append((m*i + n*j))

plot_vector(vectors)
```

When both basis vector are linearly dependant to each other so our span will be just a straight line, Means we can't make any vector except those vector whoose lie in a line. <br>
These vectors are linearly dependant to each others:
$l1 = \begin{pmatrix}
1 \\
1 
\end{pmatrix}$  $l2 = \begin{pmatrix}
2 \\
2 
\end{pmatrix}$

```python
vectors = []

l1 = np.array((1,1))
l2 = np.array((2,2))

for _ in range(300):
    m = randint(-10,10)
    n = randint(-10,10)
    vectors.append((m*l1 + n*l2))

plot_vector(vectors)
```
